### Binomial Distributions

```python
def plot_binom(n,p):
    """
    Description: Function to plot Binomial distribution
    Author: Swaroop Srisailam
    parameters:(n-number of trials,p--probability of each trial)
    """
    x = []
    y = []
    f = lambda x: (factorial(n)/(factorial(n-x) * factorial(x))) * (p**x) * ((1-p)**(n-x))
    for i in range(n+1):
        y.append(i)
        x.append(f(i))
    plt.figure(figsize = (10,6))
    plt.bar(y,x)
    plt.xlabel("trials")
    plt.ylabel("Probability")
    plt.title(f"The Binomial distribution with {n} number of Trails, prob.:{p}")
```

```python
n=int(input("Enter the number of Trials:"))
p=float(input("Enter the probability of Each Trial:"))
plot_binom(n,p)
```

### Normal Distribution


```python
def Plotting_Normal(mu,var):
    """
        Description: Function to Plot Normal Distribution
        Parameters:(mu-mean,var-variance)
        Author:Swaroop Srisailam
    """
    x= np.linspace(mu-4*var,mu+4*var)
    f= lambda x: (1/var*sqrt(2*pi))*e**((-1/2*(((x-mu)/var)**2)))
    plt.figure(figsize = (10,6))
    plt.plot(x,f(x))
    plt.scatter(mu,f(mu),color="red")
    plt.xlabel("x")
    plt.ylabel("f(x)")
    plt.title(f"Normal Distribution with mean-{mu} and varaiance-{var}")
```

```python
m=float(input("Enter the mean:"))
v=float(input("Enter the Variance:"))
Plotting_Normal(m,v)
```

### Exponential Distribution

$$f(x)=\left\{\begin{array}{cl}
\lambda e^{-\lambda x} & x\ge0\\
0 & x<0\\
\end{array}\right.$$

```python
def exponential(x,l):
    """
        Description: Function to Plot exponential Distribution
        Parameters:(x-data,l-lambda)
        Author:Swaroop Srisailam
    """
    f= lambda x,l: l*e**(-l*x)
    plt.plot(x,f(x,l))
    plt.xlabel("X")
    plt.ylabel("Probability")
        
x=np.linspace(2,4,100)
l=int(input("Enter the Lambda:"))
exponential(x,l)
    
```

## Poisson Distribution
$$ f(x) = P(X=x) = (e ^ {-\lambda} \lambda ^ x )/x! $$

```python
def fac(a):
        """
        Description:function to calculate factorial.
        Paramter:(a-[int])
        Author:Swaroop Srisailam
        """
        ans=1
        if a==0:
            return ans
        else:
            while a!=0:
                ans=ans*a
                a=a-1
            return ans

def poi(l):
    """
    Description: function to calculate the poisson distribution and plot it.
    Parameter:(l-lambda-int[])
    Author: Swaroop Srisailam
    """
    y=[]
    k=int(input("enter the x limits i.e should be greater than 0:" ))
    X=np.arange(0,k, dtype=np.float64)
    f=lambda x:exp(-l)*l/fac(x)
    for i in range(len(X)):
        y.append(f(X[i]))
    plt.plot(X,y)
    plt.title(f"Poisson Distribution with Lambda- ({l})")
    plt.xlabel("x")
    plt.ylabel("f(x)")
    plt.show()
    
#input
l=float(input("Enter the lambda value:"))
poi(l)
```

### Gamma Distribtion

$${f(x,\alpha,\lambda)={\lambda ^ \alpha  x ^{\alpha - 1}  e ^  {-\lambda x}}/\gamma (\alpha)}$$

```python
def Gamma():
    """
    Description: function to calculate the gamma function and plot it
    Author:Swaroop Srisailam
    """
    
    l=float(input("enter the lambda value: "))
    a=float(input("enter the alpha value: "))
    ans,err=quad(lambda x:(x**(a-1))*exp(-x),0,inf)
    X=np.linspace(0,10,100)
    y=[]
    l=l**a
    l_gma=l/ans
    f=lambda x: l_gma*(x**(a-1))*exp(-l*x)
    for i in range(len(X)):
        y.append(f(X[i]))
    plt.plot(X,y,label="Gamma")
    plt.xlabel("X")
    plt.ylabel('F(x,alpha,lambda)')
    plt.show()

Gamma()
```
