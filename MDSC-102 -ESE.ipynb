7. Write a program to plot the following function:
$$f(x;\theta)=(1+\theta)x^\theta, 0<x<1,\theta> -1$$

```python
#define the function
def f(x,t):
    return (1+t)*(x**t)

x=np.linspace(2,4,100)
theta=float(input("Enter theta value:"))
plt.plot(x,f(x,theta))

plt.show()
```
